# Quantitative Risk Management

- There exist risk caused by uncertainty.
- Risk management is a study to reduce risk due to uncertainty.
- Uncertainty can be quantified by measuring it.

## Quick recap: financial portfolios

A collection of "assets". These assets most likely have uncertain future returns. It includes:

- Stocks
- Bonds
- Foreign exchange holdings (forex)
- Stock options

Our challenge in risk management is to quantify risk thus making optimal investment decisions and maximize portfolio return (conditional on risk appetite)

Portfolio return is the weighted sum of individual asset returns.
