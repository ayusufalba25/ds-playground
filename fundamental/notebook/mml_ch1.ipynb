## 2.1. A System of Linear Equation

The solution of a system of linear equation is either:
- No solution
    - Contradict
- One solution: _unique solution_
- Infinitely many solution
    - Redundancy and make some variable as free variables

The geometric representation of a system of linear equation can be plotted until 3 dimension. If we have two linear equation with two variables, each equation is a line. The solution is the intersection, it can be a hyperplance, a plane, a line, or a point.

```python
from sympy import symbols, Eq, solve
from sympy.plotting import plot

# 1. Define the variables
x1, x2 = symbols('x1 x2')

# 2. Define the equations
# Equation 1: 4*x1 + 4*x2 = 5
# Equation 2: 2*x1 - 4*x2 = 1
eq1 = Eq(4*x1 + 4*x2, 5)
eq2 = Eq(2*x1 - 4*x2, 1)

# 3. Solve each equation for x2 to make them plottable (y = mx + b format)
# This returns a list, so we grab the first element [0]
y1 = solve(eq1, x2)[0]
y2 = solve(eq2, x2)[0]

# 4. Create the plot
# We plot both lines on the same graph
p = plot(y1, y2, (x1, -2, 4), show=False, legend=True)

# Optional: Customize the plot for clarity
p[0].line_color = 'blue'
p[0].label = '4x1 + 4x2 = 5'
p[1].line_color = 'red'
p[1].label = '2x1 - 4x2 = 1'
p.title = 'System of Linear Equations'
p.xlabel = 'x1'
p.ylabel = 'x2'

# 5. Show the plot
p.show()
```

## 2.2. Matrices

Matrices used to represent:
- A system of linear equation
- Linear mappings

Interesting fact: $R^{m \times n} = R^{mn}$

- Matrix multiplication consists of some dot product.
- Hadamard product: element-wise multiplication.
- Not all matrix has inverse, if it does, it is called as regular/invertible/nonsingular.
    - The inverse of a matrix is unique
    - We can compute the determinant to determine whether a matrix is invertible or not
- A matrix is symmetric if $A=A^T$
    - A symmetric matrix is also a square matrix

When using matrix to represent a system of linear equation, the pattern is: $Ax=b$ where matrix $A$ is the collective coefficient of the x.
